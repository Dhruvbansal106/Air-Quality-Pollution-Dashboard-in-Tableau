```python
# Convert 'Date' column to datetime format
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
df.info()
```

```python
# Now extract Year and Month
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df.head(30)
```

```python
# Step 4: Define the numerical columns (pollutants + AQI, including NH3)
numeric_cols = ["PM2.5", "PM10","NO","NO2","NOx","NH3","CO", "SO2","O3","Benzene","Toluene","Xylene","AQI"]  
```

```python
# Step 5: Convert these columns to numeric values
df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors="coerce")
```

```python
# Step 6: Fill missing values with city-wise monthly average
df[numeric_cols] = df.groupby(["City", "Year"])[numeric_cols].transform(lambda x: x.fillna(x.mean()))
```

```python
df.head(30)
```

#### since there are no records of PM10 and NH3 for a complete year. so we are filling PM10 by grouping the city taking average of it

```python
df[numeric_cols] = df.groupby(["City"])[numeric_cols].transform(lambda x: x.fillna(x.mean()))
```

```python
# Fill missing values with the column-wise average
df.fillna(df.mean(numeric_only=True), inplace=True)
```

```python
df.info()
```

```python
df
```

```python
# Step 7: Apply the AQI classification function
def classify_aqi(aqi):
    if aqi <= 50:
        return "Good"
    elif aqi <= 100:
        return "Moderate"
    elif aqi <= 200:
        return "Unhealthy for Sensitive Groups"
    elif aqi <= 300:
        return "Unhealthy"
    elif aqi <= 400:
        return "Very Unhealthy"
    else:
        return "Hazardous"
```

```python
df["AQI_Bucket"] = df["AQI"].apply(classify_aqi)
```
